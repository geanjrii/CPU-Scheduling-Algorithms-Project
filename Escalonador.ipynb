```python
def lerArquivo(filePath):
    file = open(filePath, 'r')
    processos = []
    for i in file:    
        processos.append([int(k) for k in i.split()])        
    file.close()      
    return processos

class Processo:
    def __init__(self, tempo_chegada, tempo_execucao, estado):        
        self.tempo_chegada = tempo_chegada 
        self.tempo_execucao = tempo_execucao
        self.estado = estado     

    def set_tempo_chegada(self, tempo_chegada):
        self.tempo_chegada = tempo_chegada

    def set_tempo_execucao(self, tempo_execucao):
        self.tempo_chegada = tempo_execucao

    def set_estado(self, estado):
        self.estado = estado

    def set_tempo_retorno(self, tempo):
        self.tempo_retorno = tempo
    
    def set_tempo_resposta(self, tempo):
        self.tempo_resposta = tempo

    def set_tempo_espera(self, tempo):
        self.tempo_espera = tempo
    
    def set_ultimo(self, ultimo):
        self.ultimo = ultimo
   
    def get_tempo_chegada(self):
        return self.tempo_chegada
        
    def get_tempo_execucao(self):
        return self.tempo_execucao

    def get_estado(self):
        return self.estado

    def get_tempo_retorno(self):
        return self.tempo_retorno
    
    def get_tempo_resposta(self):
        return self.tempo_resposta

    def get_tempo_espera(self):
        return self.tempo_espera

    def get_ultimo(self):
        return self.ultimo
    
    def sub_tempo_execucao(self, tempo_execucao):
        self.tempo_execucao -= tempo_execucao  
        
def bubbleSort(lista):
    for passnum in range(len(lista)-1,0,-1):
        for i in range(passnum):
            if lista[i].get_tempo_execucao() > lista[i+1].get_tempo_execucao():
                temp = lista[i]
                lista[i] = lista[i+1]
                lista[i+1] = temp
    return lista


def adiciona_processo(lista_processos, processos, quantidade_processos, tempo):
    for i in range(quantidade_processos):  
            if(processos[i][0] == tempo): 
                lista_processos.append(Processo(processos[i][0], processos[i][1],'preparado')) 
                

def escalonador(processos, modo):    
    lista_processos = [] 
    lista_prontos = []    
    quantidade_processos = len(processos)
    tempo_retorno = 0 
    tempo_resposta = 0 
    tempo_espera = 0
    quantum = 2
    cpu = Processo(None, None, 'finalizado')     
    tempo = 0     
        
    while(tempo < 1000):      
        
        if modo == 'RR' and len(lista_processos) == 0: 
            adiciona_processo(lista_processos, processos, quantidade_processos, tempo)  
                
        if modo != 'RR':
            adiciona_processo(lista_processos, processos, quantidade_processos, tempo)
            
        if modo=='SJF':
            lista_processos = bubbleSort(lista_processos)
        
        if(len(lista_processos) > 0 and (cpu.get_estado()=='finalizado')):                
                cpu = lista_processos.pop(0)                 
                if modo == 'RR':
                    if cpu.get_estado() == 'preparado':
                        cpu.set_tempo_resposta(tempo -cpu.get_tempo_chegada()) 
                        cpu.set_tempo_espera(cpu.get_tempo_resposta()) 
                    else:
                        cpu.set_tempo_espera((tempo -cpu.get_ultimo()) +cpu.get_tempo_espera())                    
                else:
                    cpu.set_tempo_resposta(tempo -cpu.get_tempo_chegada()) 
                    cpu.set_tempo_espera(tempo -cpu.get_tempo_chegada())                 
                cpu.set_estado('fazendo')
                
        cpu.sub_tempo_execucao(1)         
        tempo += 1       
        
        if modo == 'RR':            
            adiciona_processo(lista_processos, processos, quantidade_processos, tempo)
            quantum -= 1           
    
        if(cpu.get_tempo_execucao() == 0): 
            cpu.set_estado('finalizado') 
            cpu.set_tempo_retorno(tempo -cpu.get_tempo_chegada()) 
            lista_prontos.append(cpu) 
            if modo == 'RR':
                quantum = 2 
                
        if modo == 'RR':
            if quantum == 0:
                cpu.set_ultimo(tempo) 
                lista_processos.append(cpu) 
                cpu.set_estado('finalizado')
                quantum = 2 

        #if (cpu.get_tempo_execucao() < 0): 
        #    break                          

    for i in range(quantidade_processos):
        tempo_retorno += lista_prontos[i].get_tempo_retorno() 
        tempo_resposta += lista_prontos[i].get_tempo_resposta() 
        tempo_espera += lista_prontos[i].get_tempo_espera()  
        
    retorno = f'{modo}: {tempo_retorno/quantidade_processos} {tempo_resposta/quantidade_processos} {tempo_espera/quantidade_processos}'    
    print(retorno)
    return retorno

def gravarArquivo(processos):
    arquivo = open('EscalonadorOutput.txt', 'w')
    arquivo.write(f"{escalonador(processos, 'FCFS')}\n")
    arquivo.write(f"{escalonador(processos, 'SJF')}\n")
    arquivo.write(f"{escalonador(processos, 'RR')}\n")    

def main():
    processos = lerArquivo("Escalonador_input data.txt")
    gravarArquivo(processos)    
    
if __name__ == "__main__":
    main()
```
